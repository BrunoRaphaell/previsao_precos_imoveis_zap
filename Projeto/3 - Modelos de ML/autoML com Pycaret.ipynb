```python
import pandas as pd
import numpy as np
from pycaret.regression import *

SEED = 42
np.random.seed(seed=SEED)
```

```python
dados = pd.read_csv('../../dados/dados_OneHotEncoder.csv')
dados.head()
```

```python
exp_reg = setup(data=dados, target='price', fold_shuffle=True)
```

```python
best = compare_models(sort='rmse')
```

```python
models()
```

```python
rf = create_model('rf')
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 500, num = 100)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(3, 12, num = 3)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
turned_rf = tune_model(rf, custom_grid=random_grid, verbose=True)
```
